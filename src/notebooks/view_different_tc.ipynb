```python
import re
import os
import pandas as pd
from collections import Counter

time_controls = ['180+0', '180+2', '300+0', '300+3','600+0', '600+5', '900+10']

# Specify the path to your text file
assets_path = os.path.join(os.path.dirname(os.getcwd()), 'asset')
folder_name = 'rating_split_7_rand'
folder_path = os.path.join(assets_path, folder_name)

df = pd.DataFrame(index=time_controls)
for group in range(10):
    file_name = 'group_{}.pgn'.format(group)
    file_path = os.path.join(folder_path, file_name)

    with open(file_path, 'r') as file:
        data = file.read()

    timecontrol_matches = re.findall(r'\[TimeControl "(.*?)"\]', data)
    timecontrol_counts = Counter(timecontrol_matches)
    total_games = sum(timecontrol_counts.values())

    df[group] = [timecontrol_counts[timecontrol] for timecontrol in time_controls]
    print("Group {} Total: {} games".format(group, total_games))

df.loc['Total'] = df.sum()
```
